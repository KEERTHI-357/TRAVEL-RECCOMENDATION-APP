```python
# importing the needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Set plot style for better view
sns.set_style("whitegrid")

def clean_and_preprocess_data(
    cities_df, continents_df, countries_df, items_df, visiting_mode_df,
    regions_df, transactions_df, types_df, users_df
):
    """
    Performs data cleaning, merging, feature engineering, and preprocessing.
    """
    print("\n--- Starting Data Cleaning & Preprocessing ---")

    # Handle missing values by dropping rows where key information is missing
    transactions_df.dropna(subset=['AttractionId', 'Rating'], inplace=True)
    users_df.dropna(inplace=True)
    cities_df.dropna(subset=['CityName'], inplace=True)

    # Standardize column names for merging
    transactions_df.rename(columns={'VisitMode': 'VisitModeId'}, inplace=True)
    items_df.rename(columns={'AttractionCityId': 'CityId'}, inplace=True)

    # Merge all datasets into a single DataFrame
    merged_df = pd.merge(transactions_df, users_df, on='UserId', how='inner')
    merged_df = pd.merge(merged_df, visiting_mode_df, on='VisitModeId', how='left')
    merged_df = pd.merge(merged_df, continents_df, on='ContinentId', how='left')
    merged_df = pd.merge(merged_df, regions_df, on='RegionId', how='left')
    merged_df = pd.merge(merged_df, countries_df, on='CountryId', how='left')
    merged_df = pd.merge(merged_df, cities_df, on='CityId', how='left')
    merged_df = pd.merge(merged_df, items_df, on='AttractionId', how='left')
    merged_df = pd.merge(merged_df, types_df, on='AttractionTypeId', how='left')

    # Dropping redundant ID and address columns
    merged_df.drop(columns=[
        'VisitModeId', 'ContinentId', 'RegionId', 'CountryId', 'CityId',
        'AttractionTypeId', 'AttractionAddress', 'VisitYear', 'VisitMonth',
        'ContinentId_x', 'RegionId_x', 'CountryId_x', 'CityId_x',
        'AttractionTypeId_x', 'VisitMode_x', 'Continent_x', 'Country_x',
        'Region_x', 'Attraction_x', 'AttractionAddress_x', 'VisitMode_y',
        'Continent_y', 'Country_y', 'Region_y', 'Attraction_y',
        'AttractionAddress_y'
    ], inplace=True, errors='ignore')

    # Feature Engineering: Create a new feature for average rating per user and visit mode
    user_avg_ratings = merged_df.groupby(['UserId', 'VisitMode'])['Rating'].mean().reset_index()
    user_avg_ratings.rename(columns={'Rating': 'AvgRatingPerVisitMode'}, inplace=True)
    merged_df = pd.merge(merged_df, user_avg_ratings, on=['UserId', 'VisitMode'], how='left')

    # Preprocessing: Encoding and Normalization
    categorical_cols = ['VisitMode', 'Continent', 'Country', 'AttractionType']
    merged_df_encoded = pd.get_dummies(merged_df, columns=categorical_cols, prefix=categorical_cols)

    # Drop non-numerical columns before scaling
    numerical_df = merged_df_encoded.select_dtypes(include=[np.number])

    # Scale numerical features using MinMaxScaler
    scaler = MinMaxScaler()
    scaled_features = scaler.fit_transform(numerical_df)
    scaled_df = pd.DataFrame(scaled_features, columns=numerical_df.columns)

    return scaled_df, merged_df

def run_eda(merged_df):
    """
    Performs Exploratory Data Analysis and saves charts.
    """
    # Chart 1: User Distribution Across Continents
    continent_counts = merged_df['Continent'].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=continent_counts.index, y=continent_counts.values, palette='viridis')
    plt.title('User Distribution Across Continents', fontsize=16)
    plt.xlabel('Continent', fontsize=12)
    plt.ylabel('Number of Users', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('user_distribution_continents.png')
    plt.show()

    # Chart 2: Top 10 Most Popular Attraction Types
    attraction_popularity = merged_df.groupby('AttractionType')['Rating'].agg(['count', 'mean']).sort_values(by='count', ascending=False)
    top_10_attractions = attraction_popularity.head(10)
    plt.figure(figsize=(12, 8))
    sns.barplot(x=top_10_attractions.index, y=top_10_attractions['count'], palette='plasma')
    plt.title('Top 10 Most Popular Attraction Types by Number of Ratings', fontsize=16)
    plt.xlabel('Attraction Type', fontsize=12)
    plt.ylabel('Number of Ratings', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('top_10_attractions.png')
    plt.show()

    # Chart 3: Distribution of Ratings
    plt.figure(figsize=(8, 6))
    sns.histplot(merged_df['Rating'], bins=5, kde=True, color='skyblue')
    plt.title('Distribution of Ratings', fontsize=16)
    plt.xlabel('Rating', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.savefig('rating_distribution.png')
    plt.show()

    # Chart 4: Average Rating by Visit Mode
    visit_mode_ratings = merged_df.groupby('VisitMode')['Rating'].mean().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=visit_mode_ratings.index, y=visit_mode_ratings.values, palette='coolwarm')
    plt.title('Average Rating by Visit Mode', fontsize=16)
    plt.xlabel('Visit Mode', fontsize=12)
    plt.ylabel('Average Rating', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('average_rating_by_visit_mode.png')
    plt.show()
    # Main execution block
if __name__ == "__main__":
 try:
      transactions_df = pd.read_excel("/content/drive/MyDrive/Copy of Transaction.xlsx")
      users_df = pd.read_excel("/content/drive/MyDrive/Copy of User.xlsx")
      cities_df = pd.read_excel("/content/drive/MyDrive/Copy of City.xlsx")
      items_df = pd.read_excel("/content/drive/MyDrive/Copy of Item.xlsx")
      items_updated_df = pd.read_excel("/content/drive/MyDrive/Copy of Updated_Item.xlsx")
      types_df = pd.read_excel("/content/drive/MyDrive/Copy of Type.xlsx")
      visiting_mode_df = pd.read_excel("/content/drive/MyDrive/Copy of Mode.xlsx")
      continents_df = pd.read_excel("/content/drive/MyDrive/Copy of Continent.xlsx")
      countries_df = pd.read_excel("/content/drive/MyDrive/Copy of Country.xlsx")
      regions_df = pd.read_excel("/content/drive/MyDrive/Copy of Region.xlsx")
      print("âœ… All datasets loaded successfully.")

 except FileNotFoundError as e:
  print(f"âŒ Error loading files: {e}. Please ensure all CSV files are in the same directory.")
exit()
# Clean, preprocess, and merge data
final_df, merged_raw_df = clean_and_preprocess_data(
            cities_df, continents_df, countries_df, items_df, visiting_mode_df,
            regions_df, transactions_df, types_df, users_df
        )

        # Run EDA on the merged, but not yet encoded/scaled, DataFrame
run_eda(merged_raw_df)

        # Save the final preprocessed dataframe to a CSV file
final_df.to_csv('preprocessed_data.csv', index=False)
print("\nâœ… Data preprocessing complete. Final dataset saved to 'preprocessed_data.csv")
```

```python
# importing the needed libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# --- 1. Load the Datasets ---

try:
      transactions_df = pd.read_excel("/content/drive/MyDrive/Copy of Transaction.xlsx")
      users_df = pd.read_excel("/content/drive/MyDrive/Copy of User.xlsx")
      cities_df = pd.read_excel("/content/drive/MyDrive/Copy of City.xlsx")
      items_df = pd.read_excel("/content/drive/MyDrive/Copy of Item.xlsx")
      items_updated_df = pd.read_excel("/content/drive/MyDrive/Copy of Updated_Item.xlsx")
      types_df = pd.read_excel("/content/drive/MyDrive/Copy of Type.xlsx")
      visiting_mode_df = pd.read_excel("/content/drive/MyDrive/Copy of Mode.xlsx")
      continents_df = pd.read_excel("/content/drive/MyDrive/Copy of Continent.xlsx")
      countries_df = pd.read_excel("/content/drive/MyDrive/Copy of Country.xlsx")
      regions_df = pd.read_excel("/content/drive/MyDrive/Copy of Region.xlsx")
      print("âœ… All datasets loaded successfully.")


except FileNotFoundError as e:
    print(f"Error loading files: {e}")
    # Exit the script if a file is not found
    exit()

# --- 2. Data Preprocessing and Merging ---

# Merge transactions with items to get attraction names
merged_df = pd.merge(transactions_df, items_df, on='AttractionId', how='left')

# Drop any rows where attraction info is missing (after the left join)
merged_df.dropna(subset=['Attraction'], inplace=True)

# --- 3. Create the User-Item Matrix ---

# Pivot the merged dataframe to create a user-item matrix.
# Rows are UserIds, columns are Attraction names, and values are the ratings.
# The `fill_value=0` handles users who have not rated certain attractions.
user_item_matrix = merged_df.pivot_table(
    index='UserId',
    columns='Attraction',
    values='Rating'
).fillna(0)

print("Shape of the User-Item Matrix (Users x Attractions):")
print(user_item_matrix.shape)
print("\nFirst 5 rows of the User-Item Matrix:")
print(user_item_matrix.head())

# --- 4. Calculate User Similarity (Collaborative Filtering) ---

# Calculate the cosine similarity between users.
# This matrix will show how similar each user's rating patterns are to every other user.
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)

print("\nShape of the User Similarity Matrix:")
print(user_similarity_df.shape)
print("\nFirst 5 rows and columns of the User Similarity Matrix:")
print(user_similarity_df.iloc[:5, :5])

# --- 5. Recommendation Function ---

def get_user_recommendations(user_id, user_item_matrix, user_similarity_df, top_n=5):
    """
    Recommends attractions to a user based on collaborative filtering.

    Args:
        user_id (int): The ID of the user for whom to generate recommendations.
        user_item_matrix (pd.DataFrame): The matrix of user ratings for items.
        user_similarity_df (pd.DataFrame): The matrix of user similarity scores.
        top_n (int): The number of recommendations to return.

    Returns:
        pd.Series: A ranked list of recommended attractions with predicted ratings.
    """
    # Check if the user exists in the dataset
    if user_id not in user_similarity_df.index:
        print(f"User ID {user_id} not found in the dataset.")
        return pd.Series()

    # Get the similarity scores for the target user, sorted from highest to lowest.
    # We exclude the user's own score (which will be 1).
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index

    # Remove the target user from the list of similar users
    similar_users = similar_users[similar_users != user_id]

    # Get the items rated by the target user
    user_rated_items = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] > 0].index

    # Initialize a list to hold recommendations
    recommendations = {}

    # Iterate through similar users to find recommendations
    for similar_user in similar_users:
        # Get the items rated by the similar user
        similar_user_rated_items = user_item_matrix.loc[similar_user][user_item_matrix.loc[similar_user] > 0]

        # Find new items (not rated by the target user)
        new_items = similar_user_rated_items[~similar_user_rated_items.index.isin(user_rated_items)]

        # Add a weighted rating to our recommendations dictionary
        for item, rating in new_items.items():
            if item not in recommendations:
                recommendations[item] = []

            # The rating is weighted by the similarity score of the user
            similarity_score = user_similarity_df.loc[user_id, similar_user]
            weighted_rating = rating * similarity_score
            recommendations[item].append(weighted_rating)

    # Calculate the average weighted rating for each recommended item
    final_recommendations = {item: np.mean(ratings) for item, ratings in recommendations.items()}

    # Convert to a Series, sort by predicted rating, and return the top N
    recommended_items_series = pd.Series(final_recommendations).sort_values(ascending=False).head(top_n)
    return recommended_items_series

# --- 6. Generate and Print Recommendations ---

# Example usage: Get recommendations for a specific user ID
user_to_predict = 70456
top_recommendations = get_user_recommendations(user_to_predict, user_item_matrix, user_similarity_df, top_n=10)

print(f"\n--- Top 10 Recommended Attractions for User ID {user_to_predict} ---")

if not top_recommendations.empty:
    for i, (attraction, score) in enumerate(top_recommendations.items(), 1):
        print(f"{i}. {attraction} (Predicted Score: {score:.2f})")
else:
    print("No recommendations found for this user.")
```
